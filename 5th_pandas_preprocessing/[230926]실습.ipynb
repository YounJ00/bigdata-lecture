발행년도가 없는 출판사를 찾아, 출판사별 '발행년도 최빈값'으로 채워넣으시오
최종 df를 확인했을때 발행년도의 결측치가 0이여야함
단, 발행년도가 없을 시 랜덤년도 기입
단, 출판사가 없을 시 '출판사 없음' 입력

```python
### 발행년도가 없는 값을 뽑아 df로 만드시오
import numpy as np
no_year = df[df['발행년도'].isnull()]
no_year.head()
```

```python
#단, 출판사가 없을 시 '출판사 없음' 입력 후 df 확인
df['출판사'].fillna('출판사 없음', inplace=True)
no_year = df[df['발행년도'].isnull()]
no_year.head()
```

```python
#발행년도가 없는 출판사를 찾아, 출판사별 '발행년도 최빈값'으로 채워넣으시오
#힌트! 발행년도가 없는 출판사의 리스트를 뽑아 for loop처리!

no_year_publishers = df[df['발행년도'].isnull()]['출판사'].unique()

# 발행년도 최빈값 계산 및 채우기
for publisher in no_year_publishers:
    mode_year = df[df['출판사'] == publisher]['발행년도'].mode()
    df.loc[(df['발행년도'].isnull()), '발행년도'] = mode_year
```

```python
#df발행년도 결측치 확인
df[df['발행년도'].isnull()]
```
